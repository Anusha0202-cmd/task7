```python
import sqlite3

# Create a small SQLite database file
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()

# Create a tiny sales table
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    product TEXT,
    quantity INTEGER,
    price REAL
)
""")

# Insert sample data (for Task 7 example)
cursor.execute("INSERT INTO sales VALUES ('Coffee', 5, 20)")
cursor.execute("INSERT INTO sales VALUES ('Tea', 3, 15)")
cursor.execute("INSERT INTO sales VALUES ('Juice', 2, 12)")

conn.commit()
conn.close()

print("sales_data.db created successfully!")
```
